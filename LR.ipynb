```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("Salary_dataset.csv")
df.drop('Unnamed: 0',axis=1)
```

```python
plt.plot(df['YearsExperience'],df['Salary'],"o")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['YearsExperience']],df.Salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
m= reg.coef_
b=reg.intercept_
x=df['YearsExperience']
y=m*x+b
```

```python
reg.predict([[8.2]])
```

```python
plt.plot(df['YearsExperience'],df['Salary'],"o")
plt.plot(df['YearsExperience'],y)
```
